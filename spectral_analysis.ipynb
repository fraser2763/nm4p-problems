## Spectral Analysis

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

**FTDemo**

```python
N = 50 
f_s = 0.2123       # Sampling Frequency
tau = 1         # Sampling Interval
phi = 0        # Phase shift of sample sine wave
time = np.arange(0,N,1)
y = np.empty(50)
f = np.arange(N)/(N*tau)
function = 'o'      # 'o' for original, 'a' for sawtooth wave, ...


# Create y matrix, depending on function chosen
if function == 'o':
    for i in range(N):
        y[i] = np.sin(2*np.pi*f_s*time[i] + phi)

elif function == 'a':
    ...

# Compute Transform and decompose
Y = np.fft.fft(y)
Y_im = np.imag(Y)
Y_re = np.real(Y)

# Compute Power Spectrum
P = Y * np.conj(Y)

plt.figure(None,(14,6))
plt.subplot(1,3,1)
plt.plot(time,y)
plt.grid()
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.subplot(1,3,2)
plt.plot(f,Y_re,label='Real Part')
plt.plot(f,Y_im,'--',label='Imaginary Part')
plt.xlabel('Frequency')
plt.ylabel('Transform')
plt.legend()
plt.grid()

plt.subplot(1,3,3)
plt.semilogy(f, np.real(P))
plt.grid()
plt.xlabel('Frequency')
plt.ylabel('Power')
plt.show()
```
